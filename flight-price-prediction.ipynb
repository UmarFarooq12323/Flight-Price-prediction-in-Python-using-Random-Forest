```python
df=df.drop(columns="flight")
```

```python
df['price'].size
```

```python
def Encoder(df):
    new_df=df
    uniques=dict()
    for i in range (0,df.shape[1]):
        if (np.dtype(df.iloc[:,i])=='object'):
           uniques[df.columns.values[i]]=list(set(df.iloc[:,i]))
    ##Encoding the labels one by one 
    for name in uniques:
        labels=uniques[name]
        for i in range(0,len(labels)):
            for vals in range (0,new_df[name].size):
                if (labels[i]==df[name].values[vals]):
                    new_df[name].values[vals]=i
    new_df[list(uniques.keys())]=new_df[list(uniques.keys())].apply(pd.to_numeric)
    return new_df
        
e_df=Encoder(df)
```

```python
e_df.info()
```

```python
X,y=e_df.iloc[:,:-1],e_df.iloc[:,-1]
```

```python
X=X.drop(columns='Unnamed: 0')
```

```python
X.shape
```

```python
from sklearn.preprocessing import MinMaxScaler 
scaler1 =MinMaxScaler()
scaler2 =MinMaxScaler()
X=scaler1.fit_transform(X)
y=scaler2.fit_transform(np.array(y).reshape(-1,1))
```

```python
X_train,y_train=(X[:int(X.shape[0]*0.80),:]),y[:int(X.shape[0]*0.80)]
X_test,y_test=(X[int(X.shape[0]*0.80):,:]),y[int(X.shape[0]*0.80):]
```

Trying with simple NN

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
from tensorflow.keras import Sequential,layers
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
model=Sequential()
model.add(Dense(64,activation='relu',input_dim=9))
model.add(Dense(32,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam',metrics=['mae'],loss='mse')
```

```python
model.summary()
```

```python
h1=model.fit(x=X_train,y=y_train,epochs=20,batch_size=8)
```

```python
import matplotlib.pyplot as plt 
plt.plot(h1.history['loss'],label='Loss')
plt.legend()
plt.show()
```
